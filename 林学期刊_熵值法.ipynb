```python
import numpy as np
import pandas as pd
from EntropyMethod import EntropyMethod
data_linxue=pd.read_excel("林学期刊.xlsx",encoding='utf8') # 导入excel表中的数据
data_linxue
```

```python
# 在没有去掉网络指标的前提下
zhengxiang = data_linxue.columns.values.tolist()
del zhengxiang[0:2]
del zhengxiang[1]
zhengxiang # 获得正向指标的列表
```

```python
index = data_linxue.columns.values.tolist()
del index[0:2]
fuxiang = ['零引文率'] # 获得负向指标列表
journal_name = data_linxue['林学类期刊21'] # 包含了所有需要测量的期刊
indexs = data_linxue[index] # 包含了需要测量的指标的所有数据
em = EntropyMethod(indexs, fuxiang, zhengxiang, journal_name)
em.uniform() # 归一化后的数据
```

```python
em.calc_probability()
em.calc_entropy()
em.calc_entropy_redundancy() # 尝试计算信息效用值
```

```python
em.calc_Weight() # 计算得单项指标最终权重
```

```python
weight = em.calc_Weight().tolist()
yijizhibiao = []
yijizhibiao.append(sum(weight[0:5]))
yijizhibiao.append(sum(weight[5:10]))
yijizhibiao.append(sum(weight[10:]))
yijizhibiao = pd.Series(yijizhibiao, index=['期刊学术质量', '期刊影响力', '期刊显示度'], name='一级指标权重')
yijizhibiao # 计算一级指标权重
```
